```python
#Subfunction, do not touch
def ind(a,b):
    if a==b: return True
    return False

#for the geometric constraints, the x=-1 is here not be confuse the table number with the index in the array, consider only
#the number
def wellplaced(t):
    x=-1
    if t[x+1]==t[x+2] and t[x+3]==t[x+4]==t[x+5]: return True
    return False

#tables is a list of tables: tables=[t1,t2]
#lb is a list of lower bounds for each zone: lb=[4,5,9] means zone1 has at least 4 tables, zone 3 at least 9
#ub is a list of upper bounds for each zone: lb=[4,5,9] means zone1 has at most 4 tables, zone 3 at most 9
#n is the number of zones
def reasonable_number_tables(tables, lb, ub, n):
    zones=[0]*n
    zone=0
    for i in range(1,n+1):
        for t in tables:
            if ind(t,i)==True: zone=zone+1
        zones[i-1]=zone
        zone=0
    #print(zones)
    for i in range(n):
        if zones[i]<lb[i] or zones[i]>ub[i]: return False
    return True

#max_a is a list of upper bound for number of allergies for each zone
def reasonable_number_allergies(tables, allergies, max_a, n):
    zones=[0]*n
    allergy=0
    for i in range(1,n+1):
        for t in range(0,len(tables)):
            if ind(tables[t],i)==True: allergy=allergy+allergies[t]
        zones[i-1]=allergy
        allergy=0
    #print(zones)
    for i in range(n):
        if zones[i]>max_a[i]: return False
    return True

def allSatisfied(t,a,lb,ub,max_a,n):
    if wellplaced(t) and reasonable_number_tables(t, lb, ub, n) and reasonable_number_allergies(t, a, max_a, n):
        return True
    return False

def increment(t,n):
    if t.count(n)==len(t):
        return 0
    t.reverse()
    pos_change=0
    for i in range(len(t)):
        if t[i]!=n:
            t[i]=t[i]+1
            pos_change=i
            break;
    #print(pos_change)
    if pos_change!=0:
        t[:pos_change]=[(x+1)%n for x in t[:pos_change]]
    t.reverse()
    return 1
    
#don't touch this line
t1=t2=t3=t4=t5=t6=t7=t8=1
#here, can set the number of allergies of each table
a1=a2=a3=a4=a5=a6=a7=a8=0
#domain is from 1 to 4-1=3. For the GBD, we should use range(1,6) since we want 5 zones
domain=range(1,4)
#don't touch
n=len(domain)
t=[t1,t2,t3,t4,t5,t6,t7,t8]
a=[a1,a2,a3,a4,a5,a6,a7,a8]
#set the minimal number of tables for each section
lb=[1,1,1]
#set the maximal number of tables for each section
ub=[1,2,5]
#set the maximal number of allergies for each section
max_a=[2,2,2]
#Solving the problem using brute force, and outputs the first solution found
solutions=0
while increment(t,n)==1:
    if allSatisfied(t,a,lb,ub,max_a,n)==True: print(t); solutions=solutions+1
if solutions==0: print("No solution found")
```
